```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
from scipy.interpolate import interpn
def value_func_3d(x, y):
    return 2 * x + 3 * y
x = np.linspace(0, 4, 5)
y = np.linspace(0, 5, 6)
points = (x, y)
values = value_func_3d(*np.meshgrid(*points, indexing='ij'))
plt.scatter(*np.meshgrid(*points, indexing='ij'), c=values)
plt.colorbar()
```

```python
point = np.array([2.21, 3.12])
print(interpn(points, values, point))
```

```python
n = 60
xi = np.linspace(0, 4, 2*n)
yi = np.linspace(0, 5, n)
for i in range(2*n):
    for k in range(n):
        point = np.array([xi[i], yi[k]])
        plt.scatter(xi[i], yi[k], c=interpn(points, values, point), vmax=25, vmin=0)

plt.colorbar()
```
